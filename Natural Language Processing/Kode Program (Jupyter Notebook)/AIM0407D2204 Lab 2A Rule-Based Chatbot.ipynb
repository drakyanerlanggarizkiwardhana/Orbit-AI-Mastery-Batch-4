```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python
# definisi aturan
pairs = [
    [
        r"(hi|halo)",
        ["Halo, siapa nama kamu?"],
    ],
    [
        r"namaku (.*)",
        ["Salam kenal %1, bagaimana kabarmu hari ini?"],
    ],
    [
        r"kabarku (.*)",
        ["Senang mendengarnya. Apa yang kamu lakukan hari ini?"]
    ],
    [
        r"(.*) orbit future academy (.*)",
        ["Orbit Future Academy didirikan dengan tujuan untuk meningkatkan kualitas hidup melalui inovasi, pendidikan, dan pelatihan keterampilan"]
    ],
    [
        r"saya (.*)",
        [
            "Semangat ya!", 
            "Tetaplah seperti itu, lanjutkan!"
        ],
    ],
    [
        r"quit",
        ["Sampai jumpa! Senang mengobrol denganmu ..."],
    ]
  ]
```

```python
print("Halo, saya adalah Bot dan saya senang mengobrol. Gunakan bahasa Indonesia yang baku tanpa singkatan. Ketik 'quit' untuk mengakhiri obrolan!")

chat = Chat(pairs, reflections)
chat.converse()
```
