```python
from bertviz import head_view, model_view
from transformers import BertTokenizer, BertModel
```

# 01 Attention-head View

Memvisualisasikan attention dalam satu atau lebih kepala di lapisan tertentu dalam BERT.

```python
model_type = 'bert'
model_version = 'bert-base-multilingual-uncased'      # https://huggingface.co/bert-base-multilingual-uncased

sentence_a = 'Budi juara kelas dan dia sangat senang'
sentence_b = 'Sarah tidak juara kelas sehingga dia sedih'

tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=True)
model = BertModel.from_pretrained(model_version, output_attentions=True)

inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt', add_special_tokens=True)

input_ids = inputs['input_ids']
token_type_ids = inputs['token_type_ids']

attention = model(input_ids, token_type_ids=token_type_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0

tokens = tokenizer.convert_ids_to_tokens(input_id_list)
```

```python
head_view(attention, tokens)
```

# 02 Neuron View
Memvisualisasikan attention, serta nilai vektor query dan key di kepala attention tertentu.

* Arahkan kursor ke salah satu token di sisi kiri visualisasi untuk menyaring attention dari token itu.
* Klik ikon plus yang muncul. Ini menunjukkan vektor query, vektor key, dan komputasi perantara untuk bobot attention (biru=positif, orange=negatif).
* Arahkan kursor ke token lain di sebelah kiri untuk melihat perhitungan attention yang terkait.
* Klik pada drop-down Layer atau Head untuk mengubah layer model atau kepala (zero-indexed).



```python
from bertviz.transformers_neuron_view import BertModel, BertTokenizer
from bertviz.neuron_view import show

tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=True)
model = BertModel.from_pretrained(model_version)

show(model, model_type, tokenizer, sentence_a, sentence_b, layer=2, head=2, display_mode='light')
```

# 03 Model View
Memvisualisasikan attention secara menyeluruh di semua lapisan dan kepala.

```python
model_view(attention, tokens, display_mode='light')
```
